### Interaktives Rechenwerkzeug: Term-Umformer
Geben Sie einen algebraischen Term ein (z. B. (a**2 + b) * (a - b*c)). Das Werkzeug löst die Klammern auf und vereinfacht den Term.
Hinweis: Nutzen Sie * für Multiplikation und ** für Potenzen.

```python
%pip install sympy
```

```python
import sympy as sp

def term_werkzeug(eingabe):
    # Definition der Symbole (alle Buchstaben des Alphabets)
    buchstaben = 'abcdefghijklmnopqrstuvwxyzABCDEFGHIJKLMNOPQRSTUVWXYZ'
    syms = {char: sp.symbols(char) for char in buchstaben}
    
    try:
        # Umwandlung der Eingabe in einen SymPy-Ausdruck
        ausdruck = sp.sympify(eingabe, locals=syms)
        
        # Ausmultiplizieren und Vereinfachen
        ergebnis = sp.expand(ausdruck)
        
        print(f"Eingabe: {ausdruck}")
        print("-" * 30)
        print(f"Vereinfachter Term: {ergebnis}")
        
    except Exception as e:
        print(f"Fehler: Bitte prüfen Sie die Eingabe. (Nutzen Sie * für Multiplikation und ** für Potenzen).")
```

```python
# --- Eingabebereich ---
# Hier den Term eintragen, z.B. "(a**2 + b) * (a - b*c)"
mein_term = "(a**2 + b) * (a - b*c)"

term_werkzeug(mein_term)
```
